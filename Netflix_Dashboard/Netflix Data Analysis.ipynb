# Make a copy of the dataset

# Drop NULL values

```python
df = df.dropna()
df.shape
```

# Print first 10 values

```python
df.head(10)
```

# Convert Date Time format

```python
# Strip spaces and convert to datetime, errors become NaT
df['date_added'] = pd.to_datetime(df['date_added'].str.strip(), errors='coerce')

# Extract components safely
df['day_added'] = df['date_added'].dt.day
df['month_added'] = df['date_added'].dt.month
df['year_added'] = df['date_added'].dt.year

# (Optional) If you want to make sure they're int, and ignore NaT
df['year_added'] = df['year_added'].astype('Int64')
df['day_added'] = df['day_added'].astype('Int64')
df['month_added'] = df['month_added'].astype('Int64')
```

# Data Visualization

Type: Movie and TV Shows

```python
sns.countplot(x='type', data=netflix)
plt.gcf().set_size_inches(10,10)
plt.show()
```

# Rating of shows and movies

```python
plt.figure(figsize=(13, 13))
ax = sns.countplot(x='rating', data=netflix)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
plt.title('Rating')
plt.show()
```

# Relation between Type and Rating

```python
plt.figure(figsize=(10,8))
sns.countplot(x='rating',hue='type',data=netflix)
plt.title('Relation between Type and Rating')
plt.show()
```

# Pie-chart for the Type: Movie and TV Shows

```python
labels = ['Movie', 'TV show']
size = netflix['type'].value_counts()
colors = plt.cm.Wistia(np.linspace(0, 1, 2))
explode = [0, 0.1]
plt.rcParams['figure.figsize'] = (9, 9)
plt.pie(size,labels=labels, colors = colors, explode = explode, shadow = True, startangle = 90)
plt.title('Distribution of Type', fontsize = 25)
plt.legend()
plt.show()
```

# Pie-chart for Rating

```python
netflix['rating'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True,figsize=(10,8))
plt.show()
```
